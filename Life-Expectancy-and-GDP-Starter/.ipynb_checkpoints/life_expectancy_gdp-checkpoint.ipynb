```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Has life expectancy increased over time in the six nations?
countries = list(df['Country'].unique())
countries
```

```python
df.rename(columns={'Life expectancy at birth (years)' : 'Life expectancy'}, inplace=True)
df.head()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
n=0
for r in range(3):
    for c in range(2) :
        data_by_country = df[df['Country']==countries[n]]
        xdata=data_by_country['Year'].tolist()
        ydata=data_by_country['Life expectancy'].tolist()
        axs[r][c].set_title(f'Life Expectancy in {countries[n]}')
        axs[r][c].set_xlabel('Year')
        axs[r][c].set_ylabel('Life Expectancy')
        fig.tight_layout()
        axs[r][c].plot(xdata,ydata)
        n+=1
```

```python
#Has GDP increased over time in the six nations?
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
n=0
for r in range(3):
    for c in range(2) :
        data_by_country = df[df['Country']==countries[n]]
        xdata=data_by_country['Year'].tolist()
        ydata=data_by_country['GDP'].tolist()
        axs[r][c].set_title(f'GDP in {countries[n]}')
        axs[r][c].set_xlabel('Year')
        axs[r][c].set_ylabel('GDP')
        fig.tight_layout()
        axs[r][c].plot(xdata,ydata)
        n+=1
```

```python
# Is there a correlation between GDP and life expectancy of a country?
sns.histplot(x='Life expectancy', y='GDP', data=df)
plt.show()
```
