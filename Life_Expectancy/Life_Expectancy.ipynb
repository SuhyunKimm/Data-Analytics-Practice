```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df.head()
```

```python
df = df.rename(columns=lambda x: x.strip())
```

```python
print(f'shape = {df.shape}')
print(f'columns = {list(df.columns)}')
```

```python
df[df.isnull().any(axis=1)]
```

```python
pd.options.display.float_format='{:.2f}'.format
df.describe()
```

```python
df.describe(include='O')
```

```python
df.isnull().sum()
```

```python
df.fillna(float('inf'), inplace=True)
```

```python
# selecting columns with numeric values
df.dtypes
numeric_cols=df.select_dtypes(include=['float64','int64']).columns
len(list(numeric_cols))
```

```python
plt.figure(figsize=(20,20))

for idx, col in enumerate(numeric_cols) :
    if col != 'Life expectancy' :
        plt.subplot(7,3,idx+1)
        sns.scatterplot(x=col, y='Life expectancy', data=df)
        plt.title(f'{col} vs Life Expectancy')
plt.show()
```

```python
sns.scatterplot(x='Schooling', y='Life expectancy', data=df, hue='Status', palette='bright')
plt.show()
```
